## Load images

```python
import os

import cv2
import numpy as np

from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
N_IMAGES_TO_LOAD=50
SEGMENTED_CROPPED_DIR = os.path.abspath("img/segmented_cropped")
SEGMENTED_DIR = os.path.abspath("img/segmented")
```

```python
species_dirs = os.listdir(SEGMENTED_DIR)# os.listdir(SEGMENTED_CROPPED_DIR)

X_original = [] # the original images
Y = [] # targets

for i, specie_dir in enumerate(species_dirs): # the i iterator will be in this case, the target (each directory contains samples for a unique specie)
    specie_dir_abspath = os.path.abspath("{}/{}".format(SEGMENTED_DIR, specie_dir))
    k = 0 # reset counter for batch size
    for j, img_file_str in enumerate(os.listdir(specie_dir_abspath)):
        file_path = "{}/{}".format(specie_dir_abspath, img_file_str)
        
        if not os.path.isfile(file_path): # skip trash directories
            continue
            
        image = cv2.imread(file_path)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)# -> we don't wanna loose any information
        image = cv2.resize(image, (400, 400))
        image = image.astype(np.float32)
        image /= 255
        
        if k >= N_IMAGES_TO_LOAD:
            break
        
        k += 1
        X_original.append(image)
        Y.append(i) # append the specie "index" to Y, expected label
    #break # load only first folder

X_original = np.array(X_original)
```

## Flatten the images

We're going to obtain a matrix like

$$
    X = \begin{bmatrix}
        x_{1, 1} & x_{1, 2} & \cdots & x_{m, n}\\
        \vdots \\
        x_{m, 1} & x_{m, 2} & \cdots & x_{m, n}
    \end{bmatrix}
$$

Where

$m$: Number of SAMPLES

$n$: Number of FEATURES

```python
image_x_original_size = X_original[0].shape[0] # 400

X_flatten = X_original.reshape(-1, image_x_original_size ** 2)
print("Flattened shape: ", X_flatten.shape)
```

### Extract features

```python
#n_components=0.80 means it will return the Eigenvectors that have the 80% of the variation in the dataset
pca_images = PCA(n_components=0.9)
pca_images.fit(X_flatten)
```

```python
# get eigenvectors
eigenvectors = pca_images.transform(X_flatten)
eigen_in_our_space = pca_images.inverse_transform(eigenvectors)

fig = plt.figure(figsize=(100, 100))
for i in range(12):
    fig.add_subplot(4, 3, i + 1)
    plt.imshow(eigen_in_our_space[i].reshape(-1, 400), cmap="gray")
```

## Train

```python
import os
import numpy as np
from keras import Sequential
from keras.preprocessing.image import ImageDataGenerator
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
```

## Create model

```python
SEGMENTED_DIR = os.path.abspath("img/segmented")
image_generator = ImageDataGenerator(
    validation_split=0.2,
    rescale=1.0 / 255,
    dtype=np.float32
)

training_data = image_generator.flow_from_directory(
    SEGMENTED_DIR,
    target_size=(400, 400),
    color_mode="grayscale",
    seed=1237,
    subset="training",
    batch_size=8
    #save_to_dir=os.path.abspath("img/segmented_preprocessed")
)
```

```python
model = Sequential()
# first layer output is 100, kernel size is 3x3, input image shape is 400 w, 400 h, 1 channel
model.add(Conv2D(192, kernel_size=(3, 3), input_shape=(400, 400, 1), activation="relu"))
model.add(MaxPooling2D((2, 2)))

model.add(Conv2D(192, kernel_size=(3, 3), activation="relu"))
model.add(MaxPooling2D((2, 2)))

model.add(Conv2D(96, kernel_size=(3, 3), activation="relu"))
model.add(MaxPooling2D((2, 2)))

model.add(Conv2D(96, kernel_size=(3, 3), activation="relu"))
model.add(MaxPooling2D((2, 2)))

model.add(Flatten())
model.add(Dense(54, activation="relu"))
#model.add(Dense(125, activation="relu"))
model.add(Dense(24, activation="relu"))

model.add(Dense(6, activation='softmax'))
#model.add()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(training_data, epochs=50, batch_size=8, verbose=1)
```
